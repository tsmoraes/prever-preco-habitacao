```python
# carrega as bibliotecas para tratar os dados
import pandas as pd
import numpy as np

# bibliotecas para modelos
from sklearn.model_selection import train_test_split
import lazypredict

# O Lazy Predict ajuda a construir muitos modelos básicos sem muito código e ajuda a entender 
# quais modelos funcionam melhor sem nenhum ajuste de parâmetro.
from lazypredict.Supervised import LazyClassifier
```

```python
# carrega os dados
df = pd.read_csv("dataset/Housing.csv")
df.head()
```

```python
# seleciona as features e variavel target
X = df[['area', 'bedrooms', 'bathrooms', 'stories', 'parking']]
y = df['price']

# divide em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=999)
```

```python
X_train.head()
```

```python
clf = LazyClassifier(verbose=0, ignore_warnings=True)
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
models
```
